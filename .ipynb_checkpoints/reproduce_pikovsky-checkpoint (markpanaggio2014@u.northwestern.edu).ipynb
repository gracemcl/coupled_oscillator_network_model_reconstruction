```python
def get_phase_differences(phases,oscillator,timestep,leave_out_self=False):    
    theta_oscillator=phases[timestep,oscillator]
    if leave_out_self:
        theta_other=np.concatenate([phases[timestep,:oscillator],phases[timestep,oscillator+1:]])
    else:
        theta_other=phases
    return theta_oscillator-theta_other
```

We will format our linear system as follows:

* omega: column 1
* cos(1 dtheta):columns 2 to 2+(N-1)-1
* sin(1 dtheta):columns 2+(N-1) to 2+2(N-1)-1
* cos(2 dtheta):columns 2+2(N-1) to 2+3(N-1)-1
* sin(2 dtheta):columns 2+3(N-1) to 2+4(N-1)-1
...

```python
from scipy.sparse import lil_matrix

def generate_Ab(phases,vel,learning_params):
    M=learning_params['n_coefficients']
    N=learning_params['n_oscillators']
    T=phases.shape[0]
    cols_per_equation=1+2*M*(N-1)
    #print(M,N,T,cols_per_equation)
    A=lil_matrix((N*T, N*cols_per_equation), dtype=np.float64)
    b=vel.reshape(-1,1)
    #print(A.shape)



    for oscillator in range(N):
        startcol=oscillator*cols_per_equation
        endcol=(oscillator+1)*cols_per_equation    
        for timestep in range(T):
            startrow=(timestep)*N
            endrow=(timestep+1)*N

            ## add ones for frequencies
            for row in range(startrow,endrow):
                A[startrow+oscillator,startcol]=1

            ## add cosines and  
            for harmonic in range(0,M):

                dtheta=get_phase_differences(phases,oscillator,timestep,leave_out_self=True)
                cos_term=np.cos((harmonic+1)*dtheta)
                sin_term=np.sin((harmonic+1)*dtheta)

                col_to_store_cos=range((startcol+1+2*harmonic*(N-1)),
                                       (startcol+N+2*harmonic*(N-1)))
                col_to_store_sin=range((startcol+1+2*harmonic*(N-1))+N-1,
                                       (startcol+N+2*harmonic*(N-1))+N-1)
                #print("Cols:",startcol,endcol-1)
                #print("Rows:",startrow,endrow-1)
                #print(oscillator,timestep,harmonic)
                #print("Print current columns:",col_to_store_cos,col_to_store_sin)
                #print("Current row:",startrow+oscillator)

                A[startrow+oscillator,col_to_store_cos]=cos_term
                A[startrow+oscillator,col_to_store_sin]=sin_term
    return A,b

def unpack_x(x,learning_params):
    M=learning_params['n_coefficients']
    N=learning_params['n_oscillators']
    cols_per_equation=int(x.shape[0]/N)
    
    w=x[::cols_per_equation]
    print(w)
    n_theta=20
    th=np.linspace(-np.pi,np.pi,n_theta)
    coup_vecs=np.zeros((N-1,n_theta))
    cos_coef=np.zeros((N-1,M))
    sin_coef=np.zeros((N-1,M))
    harmonics=np.array(range(1,M+1)).reshape(-1,1)
    for oscillator in range(N):
        for neighbor in range(N-1):
            cos_coef[neighbor,:]=x[1+neighbor::cols_per_equation]
            sin_coef[neighbor,:]=x[N+neighbor::cols_per_equation]
            coup_vecs=coup_vecs+cos_coef.dot(np.cos(harmonics*th))+sin_coef.dot(np.sin(harmonics*th))
            plt.plot(th,coup_vecs[neighbor,:])
        plt.show()
        
```

```python
from scipy.sparse.linalg import lsqr
A,b=generate_Ab(phases,vel,learning_params)
sol=lsqr(A,b,damp=0.0001)
x=sol[0]
flag=sol[1] # 
print(sol[0])
```
